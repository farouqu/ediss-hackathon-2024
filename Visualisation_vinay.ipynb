```python
eeg_cols = ['ThetaFrontal', 'ThetaParietal', 'AlphaFrontal',
       'AlphaParietal', 'LowerAlphaFrontal', 'LowerAlphaParietal',
       'UpperAlphaFrontal', 'UpperAlphaParietal', 'BetaFrontal',
       'BetaParietal', 'LowerBetaFrontal', 'LowerBetaParietal',
       'UpperBetaFrontal', 'UpperBetaParietal']
```

```python
df_no_risk_t = df[df['risk_evaluation'] == 0]
df_risk_t = df[df['risk_evaluation'] != 0]
```

```python
df_no_risk_s = df2[df2['risk_evaluation'] == 0]
df_risk_s = df2[df2['risk_evaluation'] != 0]
```

```python
no_risk_dict_track = {}
for col in eeg_cols:
  no_risk_dict_track[col] = df_no_risk_t[col].sum()/len(df_no_risk_t)

risk_dict_track = {}
for col in eeg_cols:
  risk_dict_track[col] = df_risk_t[col].sum()/len(df_risk_t)
```

```python
no_risk_dict_s = {}
for col in eeg_cols:
  no_risk_dict_s[col] = df_no_risk_s[col].sum()/len(df_no_risk_s)

risk_dict_s = {}
for col in eeg_cols:
  risk_dict_s[col] = df_risk_s[col].sum()/len(df_risk_s)
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Get the labels and values for each group
labels = list(no_risk_dict_track.keys())
values1 = list(no_risk_dict_track.values())
values2 = list(risk_dict_track.values())

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width/2, values1, width, label='No Risk')
rects2 = ax.bar(x + width/2, values2, width, label='Risk')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Values')
ax.set_title('Comparison of Frequency Band Values by Brain Region (Track)')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.legend()

# Label bars with values
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
